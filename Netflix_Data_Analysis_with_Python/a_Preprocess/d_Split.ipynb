# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# load
df = pd.read_csv('../data/encodedData.csv', index_col='car_ID')
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

# Spliting

```python
from sklearn.model_selection import train_test_split
```

```python
features = df.drop(columns=['price'])
target = df.price
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.25, random_state=33)
```

```python
xtrain['price'] = ytrain
xtest['price'] = ytest
```

```python
xtrain.to_csv('../data/train.csv', index=False)
xtest.to_csv('../data/test.csv', index=False)
```
