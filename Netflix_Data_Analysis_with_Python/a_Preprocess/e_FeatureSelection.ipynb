# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load 
df = pd.read_csv('../data/train.csv')
df.head()
```

# Feature Engineering

```python
df['car_width_length'] = np.sqrt((df.carwidth + df.carlength))
df.head()
```

# Select Features

```python
features = df.drop(columns='price')
target = df.price
```

```python
from sklearn.feature_selection import mutual_info_regression, SelectKBest
```

```python
scores = mutual_info_regression(
    features,
    target,
    discrete_features='auto',
    n_neighbors=3,
    copy=True,
    random_state=33,
)
scores.shape
```

```python
scores_df = pd.DataFrame(scores, columns=['Scores'], index=df.columns[:-1]).rename({'price':'car_width_length'})
scores_df.sort_values(by='Scores', ascending=False, inplace=True)
```

```python
scores_df
```

```python
# if p_value > 0.05:
```

```python
top_features = scores_df.loc[scores_df.Scores > 0.5]
top_features
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(y=top_features.index, x=top_features.Scores, orient='h')
```

```python
selected_features = df[top_features.index]
selected_features
```

```python
selected_features.loc[:,'price'] = df.price
selected_features
```

```python
selected_features.to_csv('../data/train_v4.csv', index=False)
```
